Task1_0528. 연산메소드 이용 및 NaN 값을 0으로 대체 후 사칙연산을 수행한 후 결과를 출력하세요.

```python
import pandas as pd
import numpy as np

st1 = pd.Series({'국어':np.nan,'영어':80,'수학':90})
st2 = pd.Series({'수학':80,'국어':90})

add_s = st1.add(st2, fill_value=0)
sub_s = st1.sub(st2, fill_value=0)
mul_s = st1.mul(st2, fill_value=0)
div_s = st1.div(st2, fill_value=0)

df = pd.DataFrame([add_s, sub_s, mul_s, div_s],
                  index=['덧셈','뺄셈','곱셈','나눗셈'])
df
```

Task2_0528. 주어진 DataFrame의 N열에서 Series s값을 빼고 결과를 새로운 열 O에 저장 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "M": [15, 25, 35, 45, 55],
    "N": [100, 200, 300, 400, 500]
})
print(df,'\n')
# Series S
s = pd.Series([5, 10, 15, 20, 25])

# df의 N열 - Series S 의 결과 series 저장
new_N = df['N'].sub(s)
print(f'  N-S:\n{new_N}\n')
# df에 N-S 의 Series 삽입
# insert(location, column_name, [Scalar, Series, Array_Like], (allow_duplicates=bool))
df.insert(2, 'O', new_N)
print(df)
```

Task6_0528. df에서 두 개의 열을 더한 값이 다른 한 개의 열보다 큰 경우에는 1, 작은 경우에는 0으로 값을 정하는 새로운 열을 생성한 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [2, 3, 4, 5, 6],
    "B": [1, 2, 3, 4, 5],
    "C": [3, 4, 5, 6, 7],
    "D": [8, 9, 10, 11, 12],
    "E": [10, 20, 30, 40, 50]
})
# df의 A 컬럼과 B 컬럼을 더한 결과를 Series 객체로 저장하고 type 확인
df_ab = df[['A','B']]
df_abs = df_ab.sum(axis=1)
print(f"A 컬럼 + B 컬럼 :\n{df_abs}\n")
# A+B Series의 요소를 하나씩 꺼내서 C열의 요소들과 비교 후 value를 1 혹은 0으로 F 열 생성
for i in range(5) :
  if df_abs[i] > df.at[i, 'C'] :
    df.at[i, 'A+B | C'] = 1
  else :
    df.at[i, 'A+B | C'] = 0
print(df[['A','B','C', 'A+B | C']])
```

```python
# 강사님 풀이  = 데이터프레임끼리의 연산이 바로 되는걸 이용해서 훨씬 간결하게 코딩 가능
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [2, 3, 4, 5, 6],
    "B": [1, 2, 3, 4, 5],
    "C": [3, 4, 5, 6, 7],
    "D": [8, 9, 10, 11, 12],
    "E": [10, 20, 30, 40, 50]
})
df['A_B_greater_than_C'] = (df['A'] + df['B'] > df['C']).astype(int)

print(df)
```
