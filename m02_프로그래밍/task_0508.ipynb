Task3_0508. 다음을 수행하세요.
- 사용자로부터 문자열을 입력받습니다.
- 입력받은 문자열이 숫자로만 이루어져 있는지 확인합니다.
- 만약 숫자로만 이루어져 있다면, 그 숫자를 정수로 변환하여 100을 더한 결과를 출력합니다.
- 숫자가 아닌 문자가 포함되어 있다면, 사용자에게 숫자만 입력하라는 메시지를 출력합니다.
- 연산이 수행이 완료될 때까지 반복하고 수행이 완료되면 break 합니다.

````python
print("  ☆ 마일리지 100점을 더 드려요~ ☆\n")
while 1 :
  num_str = input("현재 마일리지 입력(enter 시 종료) : ")
  if num_str.isdigit() :
    result = int(num_str) + 100
    print(f'축하합니다. {result}점이 되었어요.')
    break
  elif num_str == '':
    break
  else :
    print("※ 숫자만 입력해주세요! ※")
print("정상 종료되었습니다.")
````

- 사용자로부터 문자열을 입력받습니다.
- 입력받은 문자열이 알파벳과 숫자로만 이루어져 있는지 확인합니다.
- 만약 조건을 만족한다면 "유효한 아이디입니다"라고 출력합니다.

````python
from datetime import datetime

def timestamp_datetime(timestamp) :
  return datetime.fromtimestamp(timestamp)
# timestamp = 1609160537.371015

ts = float(input("timestamp를 입력하세요 : "))
dt = timestamp_datetime(ts)
print(dt)
````

Task7_0508. datetime 객체를 문자열로 변환하는 사용자 함수를 작성하여 datetime 객체를 "YYYY-MM-DD HH:MM:SS" 형식의 문자열로 변환하고 출력하세요.

````python
!pip install pytz
````

````python
import datetime
from pytz import timezone

# UTC 시간을 입력하면 KOR 시간으로 변경
def datetime_to_string(dt):
  seoul_timezone = timezone('Asia/Seoul')
  dt_korea = dt.astimezone(seoul_timezone)
  return dt_korea.strftime('%Y-%m-%d %H:%M:%S')

dt = input('datetime을 입력하세요> ')
dt = datetime.datetime.strptime(dt, '%Y-%m-%d %H:%M:%S')
print(dt, type(dt))
s = datetime_to_string(dt)
print(s, type(s))
````

Task8_0508. 올해 경과된 날짜수 계산하세요.

````python
# 2024-01-01 00:00:00 -> timestamp -> 빼고 하루 몇초인지로 나누기하고 몫
# time.mktime(datetime.timetuple())
import datetime

current_date = datetime.date.today()  # 오늘 datetime.date 불러오기
start_of_year = datetime.date(current_date.year, 1, 1)  # (오늘날짜).년 불러오고, 1, 1 로 세팅해서 date 불러오기

days_passed = (current_date - start_of_year).days # datetime.date 끼리 그냥 뺄셈 가능. 뺄셈 후 .days 만 출력
print(f"올해 경과된 날짜 수 : {days_passed}일")
````

Task9_0508. 현재 요일을 "2021-6-26 오늘은 토요일입니다."와 같은 형식으로 출력하세요.

````python
import time
from datetime import datetime

wdays = {'Mon':'월요일', 'Tue':'화요일', 'Wed':'수요일', 'Thu':'목요일', 'Fri':'금요일', 'Sat':'토요일', 'Sun': '일요일'}

now  = datetime.now()
date = now.date()
wday = now.strftime("%a")

print(f'{date} 오늘은 {wdays[wday]}입니다.')
````

````python
import time

weekdays = ['월요일','화요일','수요일','목요일','금요일','토요일','일요일']
t = time.localtime()
today = f'{t.tm_year}-{t.tm_mon}-{t.tm_mday}' #tm_mday 자체가
week = weekdays[t.tm_wday]
# print('%s 오늘은 %s입니다' %(today, week))
print(f'{today} 오늘은 {week}입니다.')
````
