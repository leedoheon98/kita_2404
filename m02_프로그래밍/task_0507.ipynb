Task1_0507. 튜플 형태인 strings를 다양한 문자가 포함된 순서로 정렬하세요.

strings = ('a', 'ca', 'bar', 'aaaa', 'ababa')

```python
strings = ('a', 'ca', 'bar', 'aaaa', 'ababa')
sorted_strings = sorted(strings, key=lambda x : len(set(x)), reverse=True)
print(sorted_strings)
```

```python
lst_strings = list(strings)
lst_strings.sort(key=lambda x : len(set(x)), reverse=True)
print(lst_strings)
```

Task2_0507. eval 함수를 활용하여 a,b 변수와 연산자를 입력하면 사칙연산을 수행하는 코드를 작성하세요. (eval 사용하는 경우와 안하는 경우)

```python
# eval을 사용하는 경우
expression = input("사칙연산을 수행할 표현식 \n(a)(연산자)(b) 를 입력 : ")
result = eval(expression)
print(f'\n{result:.2f}')
```

```python
# eval을 사용하지 않는 경우
num1 = int(input("변수 a 입력 : "))
operator = input("계산할 연산자를 입력 : ")
num2 = int(input("변수 b 입력 : "))
if operator == '+':
  print(f'\n{num1} {operator} {num2} = {num1+num2}')
elif operator == '-':
  print(f'\n{num1} {operator} {num2} = {num1-num2}')
elif operator == '*':
  print(f'\n{num1} {operator} {num2} = {num1*num2}')
else :
  print(f'\n{num1} {operator} {num2} = {num1/num2:.2f}')
```

```python
# eval을 사용하지 않고 연산하는 방법 (lambda)
num = input('5+5 : ')
maths = [ '+', '-', '*', '/' ]
math = { '+':lambda a, b : a + b, '-':lambda a, b : a - b, '*':lambda a, b : a * b, '/':lambda a, b : a / b }

num_list = list(num)
for i in num_list :
  if i in maths :
    num = num.split(i)
    print(math[i](int(num[0]), int(num[1])))
```

Task4_0507. 주어진 문자열 리스트의 요소들을 대문자로 변환하세요.  
words = ["hello", "world", "python", "map"]

```python
words = ["hello", "world", "python", "map"]

for i in range(len(words)) :
  words[i] = words[i].upper()
print(words)
```

```python
# for 문이 간단할 때 리스트 컴프리헨션 항상 기억하기 !

words = ["hello", "world", "python", "map"]
print([ i.upper() for i in words ])
```

Task5_0507. 주어진 리스트에서 'p'로 시작하는 단어만 필터링하세요.

words = ["python", "is", "powerful", "programming", "language", "pandas"]

```python
words = ["python", "is", "powerful", "programming", "language", "pandas"]
filtered_words = filter(lambda x : x[0] == 'p', words)
print(list(filtered_words))
```

```python
words = ["python", "is", "powerful", "programming", "language", "pandas"]
print([ w for w in words if w.startswith('p') ])
```

Task6_0507. 내장함수를 이용해서 list = [0,1,2,3,4,5,6]에서 2를 삭제 후 출력하세요.(2가지 방법)

```python
print("1. pop() 메서드 이용하기")
lst = list(range(7))
a = 0
print(f'원본 리스트 : {lst}')
a = lst.pop(2)  # 삭제할 값의 인덱스를 입력
print(f'삭제된 값 : {a}, 변경된 리스트 : {lst}\n')

print("2. remove() 메서드 이용하기")
lst = list(range(7))
print(f'원본 리스트 : {lst}')
rm = int(input("삭제할 값을 입력 : "))
if rm in lst :
  lst.remove(rm)  # 인덱스 대신 삭제할 값을 직접 입력
  print(f'변경된 리스트 : {lst}')
else :
  print("원본 리스트 안에 있는 값을 입력!")
```

Task7_0507. 내장함수를 이용하여 동일한 개수로 이루어진 자료형을 화살표 좌측에서 우측으로 변환하세요.

[1,2,3],[4,5,6] -> [(1,4),(2,5),(3,6)]\
[1,2,3],[4,5,6] [7,8,9] -> [(1,4,7),(2,5,8),(3,6,9)]\
('abc','def') -> [('a','d'),('b','e'),('c','f')]

```python
lst1 = [1, 2, 3]
lst2 = [4, 5, 6]
lst3 = [7, 8, 9]
tp1 = ('abc', 'def')
print(f'{lst1}, {lst2}\t\t -> \t{list(zip(lst1,lst2))}')
print(f'{lst1}, {lst2}, {lst3}\t -> \t{list(zip(lst1,lst2,lst3))}')
print(f'{tp1}\t\t\t -> \t{list(zip(tp1[0], tp1[1]))}')
```

```python
lst1 = [1, 2, 3]
lst2 = [4, 5, 6]
lst3 = [7, 8, 9]
tp1 = ('abc', 'def')
print([ i for i in zip(lst1, lst2)])
print([ i for i in zip(lst1, lst2, lst3)])
print([ i for i in zip(tp1[0], tp1[1])])
```
