Task1_0430. 남녀 파트너 정해주기 프로그램(zip)

같은 수의 남녀 모임에서 파트너를 랜덤하게 정해주는 프로그램을 만들어 보세요

male = ['철수','갑돌','로미오','이몽룡','온돌']

female = ['미미','갑순','줄리엣','성춘향','평강']

```python
male = ['철수','갑돌','로미오','이몽룡','온돌']
female = ['미미','갑순','줄리엣','성춘향','평강']

rd.shuffle(male)
rd.shuffle(female)

print("*"*60)
for ml, fml in zip(male,female):
  print(ml+"♡"+fml, end=' ')
print("\n"+"*"*60)
```

```python
## 강사님 풀이
rd.shuffle(male)
rd.shuffle(female)
couples = list(zip(male, female))
for i, couple in enumerate(couples):
  print(f'커플{i+1}: {couple[0]}, {couple[1]}')
```

Task2_0430. 대문자, 소문자, 숫자를 포함하는 8자리 랜덤 비밀번호를 생성하는 프로그램을 작성하세요.*이탤릭체 텍스트*

```python
result = ''
let_num = string.ascii_letters + string.digits
rs_lst = rd.choices(let_num, k=8)
for rs in rs_lst :
  result += rs
print("8자리 랜덤 비밀번호 : ", result)
```

```python
## 강사님 풀이  (내가 놓친 부분:무조건 대문자,소문자,숫자가 한번씩은 들어가게 짜야되는것)
upper = rd.choice(string.ascii_uppercase)
lower = rd.choice(string.ascii_lowercase)
digit = rd.choice(string.digits)

characters = string.ascii_letters + string.digits
rest = ''.join(rd.choices(characters, k=5))
# 모든 요소를 하나의 문자열로 합치기
password = upper + lower + digit + rest
# 비밀번호 임의로 섞기
print(password)
password = ''.join(rd.sample(password, len(password)))
print("랜덤 비밀번호 : ", password)
```

Task3_0430. 발표자 수를 랜덤하게 출력하는 프로그램을 작성하세요.(발표자 수 입력)

kita = ['김성현','황강민','윤호준','류윤선','이상협','박지환','최환욱','서보선','김한결','김도현','김하준',
        '김도원','신현진','소지승','이범석','이현석','이명신','박윤경','이도헌','김홍준']

```python
kita = ['김성현','황강민','윤호준','류윤선','이상협','박지환','최환욱','서보선','김한결','김도현','김하준',
        '김도원','신현진','소지승','이범석','이현석','이명신','박윤경','이도헌','김홍준']
pre_num = rd.randint(1, len(kita))
pre_mem = []
count = 0
while count < pre_num :
  rc = rd.choice(kita)
  if not rc in pre_mem :
    pre_mem.append(rc)
    count += 1
  else :
    continue

pre_mem.sort()
print(f'발표자 수 : {pre_num}\n발표자 명단 : {pre_mem}')
```

```python
## 강사님 풀이
kita = ['김성현','황강민','윤호준','류윤선','이상협','박지환','최환욱','서보선','김한결','김도현','김하준',
        '김도원','신현진','소지승','이범석','이현석','이명신','박윤경','이도헌','김홍준']
num = int(input('발표자 수 입력 : '))
ans = rd.sample(kita, num)    # 중복 없이 요소를 선택 (=random.sample)
print(ans)
```

Task4_0430. 사용자로부터 숫자를 입력받아 해당 숫자의 구구단을 출력하는 프로그램을 작성하세요.

```python
mul_level = int(input("구구단의 단을 입력 : "))
print('\n'+('*'*20)+'\n')
for i in range(1, 10):
  print(f"{mul_level} x {i} = {mul_level * i}")
```

f-string 안에는 변수 뿐만 아니라 파이썬의 표현식도 올 수 있다.  
표현식(Expression)
- 하나 이상의 값, 변수, 연산자, 함수 호출 등을 조합하여 평가(evaluate)될 때 값을 생성하거나 반환하는 코드
- 표현식은 계산되어 값이 되는 모든 것을 의미

표현식의 예시:
- 수치 계산: 2 + 3는 5라는 값을 반환
- 문자열 조합: "Hello " + "World"는 "Hello World"라는 문자열을 생성
- 논리 연산: x > 5는 x의 값에 따라 True 또는 False를 반환
- 함수 호출: sum([1, 2, 3])은 리스트의 합인 6을 반환
- 조건식: x if x > 0 else -x는 x가 양수면 x를, 음수면 -x를 반환

```python
# 구구단 출력
for i in range(2, 10):
  print(f"{' '}{i}단", end='\t')
print()

for i in range(2, 10):
  print("="*6, end='\t')
print()

for i in range(1, 10):
  for k in range(2, 10):
    print(f'{k}x{i}={k*i}', end='\t')
  print()
```

Task8_0430.  1부터 10 사이의 임의의 숫자를 맞추는 게임을 만드세요. 사용자가 숫자를 맞출 때까지 입력을 계속 받으며, 정답을 맞추면 게임을 종료하세요.

```python
# 숫자 맞추기 게임
target_num = rd.randint(1, 10)
while True :
  guess = int(input("1부터 10 사이의 숫자를 맞춰보세요 : "))
  if guess > target_num :
    print("더 작은 숫자입니다!")
  elif guess < target_num :
    print("더 큰 숫자입니다!")
  else  :
    print("정답입니다 !")
    break
```

Task10_0430. 아래 사항을 반영하여 커피 자판기 프로그램을 작성하세요.
- 시나리오 : 자판기 커피 재고 5잔, 커피 1잔 가격 300원, 재고 범위내에서 300원 이상 돈을 넣으면 거스름돈과 커피를 주고 그렇지 않으면 반환하며 재고가 소진되면 안내멘트 출력
- 각 Case별 멘트 출력은 상황에 맞게 창작
- while, if ~ elif ~ else 제어문을 사용하여 작성

```python
coffee = 5
while True :
  money = int(input("돈을 주시면 커피를 드립니다."))
  if money == 300 :
    print("커피를 맛있게 드세요")
    coffee -= 1
  elif money > 300 :
    print("거스름돈 %d 원을 받으시고 커피를 맛있게 드세요"%(money-300))
    coffee -= 1
  else :
    print("금액이 부족하여 %d 원을 반환합니다." %money)
  if coffee == 0 :
    print("죄송합니다. 커피가 다 소진되었네요. 1시간 뒤에 준비해놓을게요")
    break
```
