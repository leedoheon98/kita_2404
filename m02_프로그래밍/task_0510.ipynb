```python
lst1 = ['52','273','32','문자','103']
result = []
for i in lst1 :
  try :
    result.append(int(i))
  except :
    pass
print(result)
```

- 정수를 입력하지 않으면 > '정수 아님' > '프로그램 종료'

Task5_0510. 생성자, 메소드를 모두 포함하는 클래스를 작성하고 객체 3개를 만들어서 결과를 출력하세요

```python
class Car :
  def __init__(self, name, color='white', price=2000):    # 생성자 매개변수에 (*args) 이 아닌 초기값을 설정해줌으로써
    self.name = name                                      # 해당 클래스 인스턴스를 생성할 때 name 매개변수만 필수로 주면 나머진 초기값으로 설정됨.
    self.color = color
    self.price = price

  def car_sound(self) :
    if self.price > 15000 :
      print(f'{self.name}는(은) 부와아아아')
    elif self.price > 10000 :
      print(f'{self.name}는(은) 부웅부웅')
    else :
      print(f'{self.name}는(은) 털털털털')

a = Car('아반떼')
b = Car('페라리', 'red', 17000)
c = Car('아반떼', 'black', 11000)

a.car_sound()
b.car_sound()
c.car_sound()
```
