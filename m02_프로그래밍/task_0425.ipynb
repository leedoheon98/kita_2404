Task3_0425. 주어진 리스트에서 특정 요소가 등장하는 모든 인덱스를 리스트로 만들어 출력하세요.

items = ['apple', 'banana', 'cherry', 'apple', 'cherry', 'apple']  
target = 'apple'

```python
items = ['apple', 'banana', 'cherry', 'apple', 'cherry', 'apple']
target = 'apple'
print(items, '\n', target)

index_lst = []

for i in range(0, len(items)):
  if items[i] == target :
    index_lst.append(i)

print(index_lst)
```

```python
items = ['apple', 'banana', 'cherry', 'apple', 'cherry', 'apple']
target = 'apple'

indexes = [index for index, value in enumerate(items) if value == target]

print(f"'{target}'의 모든 인덱스 : {indexes}")
```

Task4_0425. 주어진 리스트에서 연속해서 반복되는 요소만 제거하고, 결과 리스트를 반환하세요. 단, 처음 등장하는 요소는 유지해야 합니다.

예를 들어, ['a', 'a', 'b', 'c', 'c', 'c', 'd', 'e', 'e']가 입력되면, ['a', 'b', 'c', 'd', 'e']를 출력해야 합니다.

```python
print(items)
items_sorted = sorted(items)
print(items_sorted)

result_items = []
for i in range(len(items_sorted)):
  if i == 0 or items_sorted[i] != items_sorted[i-1] :
    result_items.append(items_sorted[i])

print(result_items)

# li = ['a', 'a', 'b','c','c','d','e','e']
# result = [li[0]]
# for i in li[1:]:
#   if i != result[-1]:
#     result.append(i)
# print(result)

def remove_consecutive_duplicates(items_list):
  if not items_list:
    return []

  result = [items_list[0]]

  for item in items_list[1:]:
    if item != result[-1]:
      result.append(item)

  return result
```

Task5_0425 주어진 정수 리스트와 회전 횟수 k에 대해 리스트를 오른쪽으로 k만큼 회전시킨 결과를 반환하세요. k가 리스트의 길이보다 클 수 있으며, 이 경우 k를 리스트 길이로 나눈 나머지만큼 실제 회전시키면 됩니다.

예를 들어, [1, 2, 3, 4, 5]와 k=2가 주어지면, 결과는 [4, 5, 1, 2, 3]이 되어야 합니다.

```python
def rotate_list(nums, k):
  if not nums:
    return nums
  k = k % len(nums)

  rotated = nums[-k:] + nums[:-k]
  #슬라이싱을 통해 리스트를 새로 조합
  return rotated

nums = [1,2,3,4,5]
print(nums)
k = int(input("K를 입력 : "))
if type(k) is int :
  print(rotate_list(nums,k))
else :
  print("정수만 입력해주세요.")
```

```python
def rotate_list(nums, k):
  if not nums:
    return nums
  k = k % len(nums)

  rotated = nums[k:] + nums[:k]
  #슬라이싱을 통해 리스트를 새로 조합
  return rotated

print(nums)
k = int(input("K를 입력 : "))
if type(k) is int :
  print(rotate_list(nums,k))
else :
  print("정수만 입력해주세요.")
```
