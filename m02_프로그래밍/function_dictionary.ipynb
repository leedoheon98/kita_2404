```python
class FunctionDictionary :
  def __init__(self):
    # 생성자 부분에 파이썬, 자바 함수를 담은 딕셔너리 변수 선언
    self.python_docs = {'abs':'abs()  :  abs ( number ) \n\n number 의 절댓값을 구해주는 내장함수.',
                        'str':'str()  :  str ( x ) \n\n 변수 x를 문자열로 형변환해주는 함수.',
                        'bin':'bin()  :  bin ( int ) \n\n 정수를 2 진수 문자열로 형변환.',
                        'chr':'chr()  :  chr ( int ) \n\n 정수 값에 해당하는 유니코드 문자열 형태로 반환.',
                        'map':'map()  :  map ( function , iterable , ... ) \n\n 모든 요소에 함수를 적용하고, 그 결과를 반복 가능한(iterable) 형태로 반환.',
                        'next':'next() : next ( iterator , default=None ) \n\n iterable 객체에서 iter() 함수로 반복자를 생성한 후, 이를 통해 요소를 하나씩 접근할 때 next() 함수를 사용할 수 있습니다. 더 이상 반환 요소가 없을 때는 기본적으로 StopIteration 예외를 발생.',
                        'pow':'pow()  :  pow ( x , y [, z]) \n\n거듭제곱 값을 계산하는 데 사용, 기본적으로 두 개의 인자를 받으며, 선택적으로 세 번째 인자를 받을 수 있습니다. 세 번째 인자가 주어진 경우, 계산된 거듭제곱 값에 대해 모듈로 연산을 수행.',
                        'reversed':'reversed()  :  reversed(seq) \n\n 시퀀스(예: 리스트, 튜플, 문자열 등)의 요소를 역순으로 반복하는 반복자를 반환.',
                        'round':'round()  :  round ( number , digits ) \n\n 숫자를 반올림하여 가장 가까운 정수 또는 주어진 소수점 자리까지의 숫자로 반환.',
                        'set':'set()  :  set ([ iterable ]) \n\n 중복된 요소를 포함하지 않는, 순서가 없는 컬렉션입니다. 집합은 주로 요소의 존재 여부를 효율적으로 확인하거나 중복을 제거할 때 사용.',
                        'slice':'slice()  :  slice ( start , stop [, step]) \n\n 시퀀스(리스트, 튜플, 문자열 등)를 부분적으로 접근하기 위한 슬라이스 객체를 생성하는 데 사용.',
                        'sorted':'sorted()  :  sorted ( iterable , * , key=None , reverse=False ) \n\n *reverse: True로 설정하면, 리스트가 역순(내림차순)으로 정렬됩니다. 기본값은 False이며, 오름차순으로 정렬합니다. \n 어떤 반복 가능한(iterable) 객체의 모든 요소를 새로운 정렬된 리스트로 반환하는 데 사용.',
                        'zip':'zip()  :  zip ( *iterables ) \n\n *iterables: 하나 이상의 이터러블 객체(예: 리스트, 튜플, 문자열 등). \n 여러 개의 이터러블(iterables, 예: 리스트, 튜플)을 인자로 받아, 동일한 인덱스에 있는 요소들을 묶어서 튜플로 구성된 이터레이터(iterator)를 생성.'
                       }

    self.java_docs = {'repeat':'repeat()  :  String.repeat ( int count ) \n\n해당 문자열을 주어진 횟수(count) 만큼 반복하여 새로운 문자열을 반환한다. \n count가 0일 경우에는 빈문자열("")을 반환한다.',
                      'length':'length()  :  length ( ) \n\n 길이를 반환한다. 위와 같이 String 클래스에 속한 메소드',
                      'indexOf':'indexOf()  :  indexOf ( str ) \n\n 특정 문자열이 대상 문자열의 몇 번째 인덱스에 위치하는지 반환 \n 특정 문자열이 없을 경우에는 -1을 리턴',
                      'replace':'replace()  :  replace(beforeStr, afterStr) \n\n 특정 문자열을 새로운 문자열로 치환'}

  def show_docs(self):    # " FunctionDictionary "에 들어있는 함수를 모두 보여줌.

    print(f"{' 파이썬 함수':=>30} {'='*30}\n")
    # 사전 형태로 보여주기 위해 키 값으로 a-z 정렬.
    sorted_keys = sorted(self.python_docs.keys())
    sorted_values = [ self.python_docs[key] for key in sorted_keys ]
    for p_value in sorted_values :
      print("|", p_value,'\n')
    print("\n")

    print(f"{' 자바 함수':=>31} {'='*30}\n")
    sorted_keys = sorted(self.java_docs.keys())
    sorted_values = [ self.java_docs[key] for key in sorted_keys ]
    for j_value in self.java_docs.values() :
      print("|", j_value,'\n')
    print("\n")

    print("="*70)

  def help_docs(self):   # 실질적으로 사용자가 항목을 찾기위해 실행되는 사용자 함수

    while True :
      input_language = input("1. 파이썬  2. 자바  0. 종료  ▶ ")

      if input_language == '1':
        print(f"{' 파이썬 함수':=>30} {'='*30}\n")
        for p_key in self.python_docs :
          print(p_key, end=' ')
        print("\n")
        input_doc = input("궁금한 메서드 입력 ▶ ")
        if input_doc in self.python_docs.keys() :
          result = self.python_docs[input_doc].splitlines()
          print("")
          for rs in result :
            print("|", rs)
          print(f"{'='*70}\n")
        else :
          print("※ 입력 메서드가 잘못 입력되었습니다. 언어를 다시 선택해주세요. \n")

      elif input_language == '2' :
        print(f"{' 자바 함수':=>31} {'='*30}\n")
        for j_key in self.java_docs :
          print(j_key, end=' ')
        print("\n")
        input_doc = input("궁금한 메서드 입력 ▶ ")
        if input_doc in self.java_docs.keys() :
          result = self.java_docs[input_doc].splitlines()
          print("")
          for rs in result :
           print("|", rs)
          print(f"{'='*70}\n")
        else :
          print("※ 입력 메서드가 잘못 입력되었습니다. 언어를 다시 선택해주세요. \n")

      elif input_language == '0':
        print("☆  함수 사전을 종료합니다 !  ☆")
        break

      else :
        print("※ 해당 언어는 등록되어 있지 않습니다. 언어를 다시 선택해주세요. \n")
        continue


  def set_docs(self, password):
    if password == 'kita2404':
      self.show_docs()
      while True :
        f_language = input("1. 파이썬  2. 자바  0. 종료 : ")

        if f_language == '1' or f_language == '2' :
          f_mode = input("a. 추가  d. 삭제 : ").lower()

          if f_mode == 'a' :

            f_name = input("사전에 추가할 함수명을 입력 해주세요 : ")
            f_discript = input("해당 함수를 설명해주세요 : ")

  # 딕셔너리에 현재 등록되어 있지않은 함수명이면 추가
            if f_language == '1' and not f_name in self.python_docs.keys() :
              self.python_docs[f_name] = f_discript
              print('성공적으로 완료되었습니다.\n')
              for p_key in self.python_docs :
                print(p_key, end=' ')
              print("\n")

            elif f_language == '2' and not f_name in self.java_docs.keys() :
              self.java_docs[f_name] = f_discript
              print('성공적으로 완료되었습니다.\n')
              for j_key in self.java_docs :
                print(j_key, end=' ')
              print("\n")

            else :
              print("※ 해당 함수는 이미 등록되어있습니다.")

          elif f_mode == 'd' :
            f_name = input("삭제할 함수명을 입력 해주세요 : ")
  # 딕셔너리에 현재 등록되어있는 동일한 함수명이 있으면 삭제
            if f_language == '1' and f_name in self.python_docs.keys() :
              del(self.python_docs[f_name])
              print(f'성공적으로 완료되었습니다.\n')
              for p_key in self.python_docs :
                print(p_key, end=' ')
              print("\n")

            elif f_language == '2' and f_name in self.java_docs.keys() :
              del(self.java_docs[f_name])
              print(f'성공적으로 완료되었습니다.\n')
              for j_key in self.java_docs :
                print(j_key, end=' ')
              print("\n")

            else :
              print("※ 해당 함수는 등록되어 있지 않습니다.")

          else :
            print("※ 현재 추가와 삭제만 할 수 있습니다.")


        elif f_language == '0' :
          print("set doc 을 종료합니다.")
          break

        else :
          print("※ 등록되어있는 언어 중에 골라주세요. ")
    else :
      print("※ 해당 메서드는 관리자말고는 불러올 수 없습니다.")


  def main(self):
    self.show_docs()
    self.help_docs()
```
