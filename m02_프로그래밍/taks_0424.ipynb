Q. 주어진 문자열 리스트에서 유효한 Python 변수 이름만을 추출하여 반환하는 함수를 작성하세요.

```python
identifiers = ["var1", "2things", "variable_name", "time!"]
```

* re 모듈의 .match 메서드 이용
* *string*.isidentifier() 메서드 이용
* identifiers[1:] 슬라이싱 후 isalnum() 메서드 활용



```python
import re
import keyword

def find_valid_identifiers(identifiers):
    valid_identifiers = []
    for identifier in identifiers:
        # 변수 이름이 숫자로 시작하는지 확인하고 continue
        if not identifier[0].isalpha():
            continue
        # 변수 이름에 특수문자가 포함되는지 확인하고 continue
        if not re.match('^\w+$', identifier):
            continue
        # 변수 이름에 예약어가 포함되는지 확인하고 continue
        if keyword.iskeyword(identifier):
            continue
        # 유효한 변수명 만을 valid_identifiers 에 추가
        valid_identifiers.append(identifier)
    return valid_identifiers

identifiers = ["var1", "2things", "variable_name", "time!"]
valid_identifiers = find_valid_identifiers(identifiers)
print(valid_identifiers)
```
