Task2_0426. 두 개의 리스트에서 공통으로 나타나는 요소들을 찾아 리스트로 반환하는 프로그램을 작성하세요.

예시 입력: list1 = [1, 2, 3, 4, 5], list2 = [4, 5, 6, 7, 8]

예시 출력: [4, 5]

```python
def list_intersection(list1, list2):
  set1, set2 = set(), set()
  set1 = set(list1)
  set2 = set(list2)
  set_inter = set1.intersection(set2)
  result = list(set_inter)
  return result

list1 = [1,2,3,4,5]
list2 = [4,5,6,7,8]
result = list_intersection(list1, list2)
print(result)
```

Task3_0426. 주어진 숫자 리스트에서 특정 값보다 큰 요소가 하나라도 존재하는지 검사하고, 그 결과를 불 값으로 반환하는 함수를 작성하세요.

- numbers = [1, 2, 3, 10, 20]
- threshold = 15

```python
def compare_num(numbers, cnum):
  cn_result = 0
  for num in numbers:
    if cnum < num :
      cn_result = num
  return bool(cn_result)


numbers = [1,2,3,10,20]
threshold = int(input("비교할 숫자를 입력 : "))
print("리스트에 하나라도 큰 요소가 있습니까?\nAnswer is : ", compare_num(numbers, threshold))
print("type is ",type(compare_num(numbers, threshold)))
```

Task4_0426. 주어진 문자열이 특정 문자열을 포함하는지 확인하고 결과를 불 값으로 반환하는 함수를 작성하세요.

- text = "hello world"
- substring = "world"

```python
def check_text(text, word):
  if not word in text :
    return True
  else :
    return False

text = "hello world"
substring = "world"
print(check_text(text, substring))
print("type is ",type(check_text(text, substring)))
```

Task5_0426. 주어진 연도가 윤년인지 판별하는 함수를 작성하세요. 윤년은 다음의 조건을 만족해야 합니다:
- 4로 나누어 떨어진다.
- 100으로 나누어 떨어지지 않거나, 400으로 나누어 떨어지면 윤년이다.

year = 2020

True

```python
def leapyear(year):
  if year % 4 == 0 :
    if year % 100 != 0 or year % 400 == 0 :
      return True
  else :
    return False

year = int(input("연도를 입력해주세요 : "))
print(leapyear(year))
print("type is ",type(leapyear(year)))
```

Task6_0426. Calculator 클래스를 작성하고 4칙연산을 수행하는 객체 4개를 작성하여 결과를 출력하세요.

```python
class Calculator:
  def add(self,a,b) :
    return a+b
  def sub(self,a,b) :
    return a-b
  def mul(self,a,b) :
    return a*b
  def div(self,a,b) :
    if b == 0 :
      print("0으로는 나눌 수 없습니다.")
    else :
      return round(a/b,2)

calculator = Calculator()
num1 = int(input("첫 번째 숫자 : "))
num2 = int(input("두 번째 숫자 : "))

print(calculator.add(num1,num2))
print(calculator.sub(num1,num2))
print(calculator.mul(num1,num2))
print(calculator.div(num1,num2))
```

Task7_0426. 다음 과제를 수행하세요.
- 사용자로 부터 텍스트를 입력 받는다.
- 문자을 단어 단위로 분리 : split()
- 단어의 빈도수를 저장할 딕셔너리를 생성
- 각 단어의 빈도 수를 계산(for 문 / if else문)
- 결과 출력

[ 예시 ]

문장을 입력하세요: I love apple. I love orange. Apple is tasty

{'i': 2, 'love': 2, 'apple.': 1, 'orange.': 1, 'apple': 1, 'is': 1, 'tasty': 1}

```python
sentence = input("문장을 입력하세요 : ")
sentence = sentence.lower()
spl_sen = sentence.split()
w_count = {}
for ss in spl_sen :
  if ss in w_count :
    w_count[ss] += 1
  else :
    w_count[ss] = 1
print(w_count)
```
